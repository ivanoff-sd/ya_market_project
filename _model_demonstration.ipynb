## загружаем необходимые данные

```python
import pandas as pd
import numpy as np
import pickle
import shutup
```

```python
# импортируем кастомные классы
from class_predictor import predictor
from class_packer import packer
```

```python
shutup.please()
```

```python
carton = pd.read_csv('carton_full_features.csv')
sku = pd.read_csv('full_sku_data.csv')
```

```python
# откроем модели
with open('model_big.pcl', 'rb') as f:

    model_big = pickle.load(f)
    
with open('model_small.pcl', 'rb') as f:

    model_small = pickle.load(f)
```

## Создадим списки SKU для проверки алгоритма

```python
sample_1 = sku['sku'].sample(n=100)
sample_2 = sku['sku'].sample(n=150)
sample_3 = sku['sku'].sample(n=250)
sample_4 = sku['sku'].sample(n=1)
sample_5 = sku['sku'].sample(n=5)

# сэмплы возьмем как с большим количеством товаров, так и с малым.
```

## Используем класс predictor

```python
# здесь можно ознакомиться с аргументами, которые принимает класс при инициализации
pred = predictor(model_large=model_big, model_small=model_small, sku=sku, carton=carton)
```

```python
predictions_1 = pred.predict(sample_1)
predictions_2 = pred.predict(sample_2)
predictions_3 = pred.predict(sample_3)
predictions_4 = pred.predict(sample_4)
predictions_5 = pred.predict(sample_5)
```

```python
for dataframe in [predictions_1, predictions_2, predictions_3, predictions_4, predictions_5]:
    n=1
    print('predictions_',n)
    n+=1
    display(dataframe.head())
    display(dataframe.info())
    
```

Как видно выше, данные обогатились сведениями из датафрейма sku и получили столбец prediction. Теперь их можно скормить упаковщику.

## Используем класс packer

Этот класс принимает результат работы предыдущего класса и carton при инициализации, а ответ выдает по команде packer.pack()

```python
pc = packer(predictions_1, carton)
result_1 = pc.pack()
pc = packer(predictions_2, carton)
result_2 = pc.pack()
pc = packer(predictions_3, carton)
result_3 = pc.pack()
pc = packer(predictions_4, carton)
result_4 = pc.pack()
pc = packer(predictions_5, carton)
result_5 = pc.pack()
```

```python
for df in [result_1, result_2, result_3, result_4, result_5]:
    display(df)
```

Инструкции по упаковке получены.

Далее, чтоб убедиться, что эти данные соответствуют действительности, предлагаю сделать следующее:
    
- посчитать sku в каждом результате и сравнить с количеством поданных на вход.
- для небольших файлов ознакомиться с данными по коробкам и товарам, чтоб убедиться в том, что они друг другу подходят.

```python
n=1
for df in [result_1, result_2, result_3, result_4, result_5]:
    
    print('result number', n)
    n+=1
    item_list = []
    for x in df['goods']:
        item_list.append(len(x))
    print(item_list)
    print(sum(item_list))
```
